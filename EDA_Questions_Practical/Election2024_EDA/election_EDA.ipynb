```python

import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv('election2024.csv')

# 1. Age distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=30, kde=True)
plt.title('Age Distribution of Survey Respondents')
plt.xlabel('Age')
plt.ylabel('Count')
plt.savefig('age_distribution.png')
plt.show()
```

```python
# 2. Voting intention by candidate
plt.figure(figsize=(12, 6))
candidate_counts = df['candidate'].value_counts()
sns.barplot(x=candidate_counts.index, y=candidate_counts.values)
plt.title('Voting Intention by Candidate')
plt.xlabel('Candidate')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('voting_intention.png')
plt.show()
```

```python
# 3. Relationship between age and internet usage
plt.figure(figsize=(12, 6))
sns.boxplot(x='internet_usage', y='age', data=df, order=['do not use', 'less than 1 hour a day', '1-2 hours a day', '2-4 hours a day', 'over 4 hours a day'])
plt.title('Age Distribution by Internet Usage')
plt.xlabel('Internet Usage')
plt.ylabel('Age')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('age_internet_usage.png')
plt.show()

print("Graphs have been generated and saved as PNG files.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv('election2024.csv')
```

```python
# 1. Relationship between education and candidate preference
plt.figure(figsize=(12, 6))
sns.countplot(x='education', hue='candidate', data=df)
plt.title('Candidate Preference by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Candidate', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig('education_candidate.png')
plt.show()
```

```python
# 2. Voting intention by age group

df['age_group'] = pd.cut(df['age'], bins=[0, 25, 35, 45, 55, 65, 100], labels=['18-25', '26-35', '36-45', '46-55', '56-65', '65+'])
plt.figure(figsize=(10, 6))
sns.countplot(x='age_group', hue='will_vote', data=df)
plt.title('Voting Intention by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.legend(title='Will Vote', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig('voting_intention_age.png')
plt.show()
```

```python
# 3. Correlation Heatmap

plt.figure(figsize=(10, 8))
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.savefig('correlation_heatmap.png')
plt.show()
```

```python
# 2. Internet Usage by Age Group

plt.figure(figsize=(12, 6))
sns.countplot(x='age_group', hue='internet_usage', data=df)
plt.title('Internet Usage by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.legend(title='Internet Usage', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig('internet_usage_age.png')
plt.show()
```

```python
# 1. Education Level by Candidate Preference

plt.figure(figsize=(12, 6))
sns.countplot(x='education', hue='candidate', data=df)
plt.title('Education Level by Candidate Preference')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.legend(title='Candidate', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('education_candidate.png')
plt.show()
```

```python
# 2. Income Level by Voting Intention

plt.figure(figsize=(12, 6))
sns.countplot(x='income', hue='will_vote', data=df)
plt.title('Income Level by Voting Intention')
plt.xlabel('Income Level')
plt.ylabel('Count')
plt.legend(title='Will Vote', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('income_voting_intention.png')
plt.show()
```

```python
# 3. Employment Status by Federal District

plt.figure(figsize=(14, 6))
sns.countplot(x='federal_district', hue='employment', data=df)
plt.title('Employment Status by Federal District')
plt.xlabel('Federal District')
plt.ylabel('Count')
plt.legend(title='Employment', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('employment_district.png')
plt.show()
```

```python
# Internet Usage by Education Level

plt.figure(figsize=(14, 8))
sns.countplot(data=df, x='education', hue='internet_usage')
plt.title('Internet Usage by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Internet Usage')
plt.show()
```
